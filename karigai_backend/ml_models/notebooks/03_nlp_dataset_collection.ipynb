```python
import sys
sys.path.append('../utils')

import os
import json
import pandas as pd
import numpy as np
from pathlib import Path
from nlp_dataset_preparation import (
    IntentDatasetPreparator,
    TranslationDatasetPreparator,
    KnowledgeDatasetPreparator,
    SchemeDatasetPreparator
)
```

## 1. Intent Recognition Dataset

Create datasets for classifying user intents and extracting entities.

```python
# Prepare intent recognition datasets
intent_preparator = IntentDatasetPreparator('../data/processed/nlp/intent')
intent_preparator.prepare_datasets(num_examples=10000)

print("Intent recognition datasets created!")
```

```python
# Load and inspect intent dataset
with open('../data/processed/nlp/intent/intent_train.json', 'r', encoding='utf-8') as f:
    intent_data = json.load(f)

print(f"Total training examples: {len(intent_data)}")
print("\nSample examples:")
for i in range(5):
    example = intent_data[i]
    print(f"\nText: {example['text']}")
    print(f"Intent: {example['intent']}")
    print(f"Entities: {example['entities']}")
    print(f"Language: {example['language']}")
```

```python
# Analyze intent distribution
intent_counts = {}
for example in intent_data:
    intent = example['intent']
    intent_counts[intent] = intent_counts.get(intent, 0) + 1

print("Intent distribution:")
for intent, count in sorted(intent_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"{intent}: {count} ({count/len(intent_data)*100:.1f}%)")
```

## 2. Translation Dataset

Create parallel corpora for translation between Indian languages.

```python
# Prepare translation datasets
translation_preparator = TranslationDatasetPreparator('../data/processed/nlp/translation')
translation_preparator.prepare_datasets(num_pairs=50000)

print("Translation datasets created!")
```

```python
# Load and inspect translation dataset
with open('../data/processed/nlp/translation/translation_train.json', 'r', encoding='utf-8') as f:
    translation_data = json.load(f)

print(f"Total translation pairs: {len(translation_data)}")
print("\nSample pairs:")
for i in range(5):
    pair = translation_data[i]
    print(f"\nSource ({pair['source_lang']}): {pair['source']}")
    print(f"Target ({pair['target_lang']}): {pair['target']}")
    print(f"Register: {pair['register']}")
```

## 3. Knowledge Retrieval Dataset

Create question-answer pairs for troubleshooting and knowledge retrieval.

```python
# Prepare knowledge retrieval datasets
knowledge_preparator = KnowledgeDatasetPreparator('../data/processed/nlp/knowledge')
knowledge_preparator.prepare_datasets(num_pairs=5000)

print("Knowledge retrieval datasets created!")
```

```python
# Load and inspect knowledge dataset
with open('../data/processed/nlp/knowledge/knowledge_train.json', 'r', encoding='utf-8') as f:
    knowledge_data = json.load(f)

print(f"Total QA pairs: {len(knowledge_data)}")
print("\nSample QA pairs:")
for i in range(5):
    qa = knowledge_data[i]
    print(f"\nQuestion: {qa['question']}")
    print(f"Context: {qa['context']}")
    print(f"Answer: {qa['answer']}")
    print(f"Category: {qa['category']}")
```

## 4. Government Scheme Dataset

Create scheme information and user profiles for matching.

```python
# Prepare government scheme datasets
scheme_preparator = SchemeDatasetPreparator('../data/processed/nlp/schemes')
scheme_preparator.prepare_datasets(num_schemes=500, num_profiles=5000)

print("Government scheme datasets created!")
```

```python
# Load and inspect scheme dataset
with open('../data/processed/nlp/schemes/schemes.json', 'r', encoding='utf-8') as f:
    schemes = json.load(f)

print(f"Total schemes: {len(schemes)}")
print("\nSample schemes:")
for i in range(3):
    scheme = schemes[i]
    print(f"\nScheme ID: {scheme['scheme_id']}")
    print(f"Name: {scheme['name']}")
    print(f"Description: {scheme['description']}")
    print(f"Eligibility: {scheme['eligibility']}")
    print(f"Benefits: {scheme['benefits']}")
    print(f"Documents: {scheme['documents']}")
```

```python
# Load and inspect user profiles
with open('../data/processed/nlp/schemes/profiles_train.json', 'r', encoding='utf-8') as f:
    profiles = json.load(f)

print(f"Total user profiles: {len(profiles)}")
print("\nSample profiles:")
for i in range(3):
    profile = profiles[i]
    print(f"\nUser ID: {profile['user_id']}")
    print(f"Age: {profile['age']}")
    print(f"Income: {profile['income']}")
    print(f"Trade: {profile['trade']}")
    print(f"State: {profile['state']}")
    print(f"Education: {profile['education']}")
```

## 5. Dataset Statistics

Analyze overall dataset statistics.

```python
# Calculate dataset statistics
stats = {
    'Intent Recognition': {
        'Train': len(intent_data),
        'Classes': len(set(e['intent'] for e in intent_data)),
        'Languages': len(set(e['language'] for e in intent_data))
    },
    'Translation': {
        'Train': len(translation_data),
        'Language Pairs': len(set((p['source_lang'], p['target_lang']) for p in translation_data)),
        'Registers': len(set(p['register'] for p in translation_data))
    },
    'Knowledge Retrieval': {
        'Train': len(knowledge_data),
        'Categories': len(set(qa['category'] for qa in knowledge_data))
    },
    'Government Schemes': {
        'Schemes': len(schemes),
        'User Profiles': len(profiles)
    }
}

print("\n=== Dataset Statistics ===")
for dataset, metrics in stats.items():
    print(f"\n{dataset}:")
    for metric, value in metrics.items():
        print(f"  {metric}: {value}")
```
